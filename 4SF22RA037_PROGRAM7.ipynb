Program 9 : write an intermediate programming exercise with python code that demonstrates fundamentals of data visualization using the
Matplotlib library. Create histogram and scatterplot. use Cars.csv dataset from kaggle.


```python
import pandas as pd
import matplotlib.pyplot as plt
# Step 1: Load the dataset
data = pd.read_csv('/content/Car (1).csv')
# Step 2: Clean the data
data.dropna(inplace=True) # Remove rows with missing values
data = data[data['Price_in_thousands'] < 50000] # Remove outliers
# Step 3: Visualize the distribution of car prices using a histogram
plt.figure(figsize=(10, 6))
plt.hist(data['Price_in_thousands'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Car Prices')
plt.xlabel('Price_in_thousands')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
# Step 4: Explore the relationship between car price and horsepower using a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Horsepower'], data['Price_in_thousands'], color='orange', alpha=0.5)
plt.title('Car Price vs. Horesepower')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

Conclusion
Loading the Dataset: The code imports the necessary libraries, particularly pandas for data manipulation and matplotlib for visualization.
Then it loads the dataset from a CSV file named 'Car.csv' into a pandas DataFrame.
Data Cleaning: The code cleans the data by removing rows with missing values (dropna) and filtering out outliers in the
'Price_in_thousands' column to ensure the data's quality.
Visualizing Car Price Distribution: It generates a histogram to visualize the distribution of car prices. This provides insights into how prices
are distributed among the cars in the dataset.
Exploring Relationship Between Car Price and Horsepower: A scatter plot is created to explore the relationship between car price and
horsepower. This allows for understanding if there's any correlation or pattern between these two variables.

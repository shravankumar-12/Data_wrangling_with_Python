Program 4:** perform exploratory data analysis using Seaborn and Matplotlib libraries. Create
differrent types of plots and visualizations to understand data distribution and relationships. use a
publicly available dataset


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
# Load the Iris dataset
iris = sns.load_dataset("iris")
# Display the first few rows of the dataset
print(iris.head())
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```

** Boxplot:** Boxplots can be used to visualize the distribution of a single continuous variable or the
distribution of a continuous variable within different categories.

```python
sns.boxplot(x="species", y="sepal_length", data=iris)
plt.show()
```

** Histogram:** Histograms are useful for visualizing the distribution of a single continuous
variable.


```python
sns.histplot(iris["petal_length"], bins=20, kde=True)
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.title("Distribution of Petal Length")
plt.show()
```

** Violin plot:** Violin plots are similar to boxplots but also display the probability density of the
data at different values.

```python
sns.violinplot(x="species", y="petal_width", data=iris)
plt.show()
```

Correlation Heatmap: A correlation heatmap can be useful to visualize the correlation between
different variables in the dataset

```python
numerical_iris = iris.select_dtypes(include=['float', 'int'])
correlation_matrix = numerical_iris.corr()
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

Barplot: Barplots are helpful for visualizing the distribution of a categorical variable or the
relationship between a categorical variable and a continuous variable.


```python
sns.barplot(x="species", y="sepal_width", data=iris)
plt.xlabel("Species")
plt.ylabel("Sepal Width")
plt.title("Mean Sepal Width by Species")
plt.show()
```

Scatter plot: Scatter plots are useful for visualizing the relationship between two continuous
variables.


```python
sns.scatterplot(x="petal_length", y="petal_width", data=iris, hue="species")
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Length vs Petal Width")
plt.show()
```

Jointplot: Jointplots can be used to visualize the relationship between two variables along with
their individual distributions.

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind="scatter")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title(" Jointplot of Sepal Length and Sepal Width")
plt.show()
```

PairGrid: PairGrid allows you to create a grid of subplots for pairwise relationships, and it offers
exibility in customizing each subplot.

```python
g = sns.PairGrid(iris, hue="species")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
plt.show()
```

** Strip plot:** Strip plots are similar to scatter plots but are especially useful for visualizing the
distribution of a continuous variable across different categories.


```python
sns.stripplot(x="species", y="petal_length", data=iris, jitter=True)
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.title("Strip Plot of Petal Length by Species")
plt.show()
```

In conclusion, exploratory data analysis using Seaborn and Matplotlib revealed valuable insights into the dataset's distribution, relationships, and potential outliers, setting a solid foundation for further in-depth analysis and modeling.
